```python
#Read Data
import json
import pandas as pd
from pandas.io.json import json_normalize
import pandas as pd
import io

file_path = 'image_review_all.json'


with io.open(file_path,'r',encoding='utf-8') as file:
    content = file.readlines()
extracted_data = []
for entry in content:
    data = json.loads(entry)
    #print(data)
    business_id = data['business_id']
    user_id = data['user_id']
    rating = data['rating']
    review_text = data['review_text']
    words = review_text.split()
    review_length = len(words)
    pic_count = len(data['pics'])
    
    extracted_data.append({
        "business_id": business_id,
        "user_id": user_id,
        "rating": rating,
        "review_text": review_text,
        "review_length": review_length,
        "pic_count": pic_count
    })

# Creating DataFrame
df = pd.DataFrame(extracted_data)
df.head()  # Displaying the first few rows of the DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Basic statistics
rating_stats = df['rating'].describe()

# Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Number of reviews per user
reviews_per_user = df['user_id'].value_counts()

# Plotting the number of reviews per user
plt.figure(figsize=(10, 6))
sns.histplot(reviews_per_user, bins=30, kde=False)
plt.title('Number of Reviews per User')
plt.xlabel('Number of Reviews')
plt.ylabel('Count of Users')
plt.show()

rating_stats, reviews_per_user.describe()
```

```python
# Using the previously created DataFrame 'df_main' to calculate the frequency of reviews per business
business_review_counts = df['business_id'].value_counts()

# Plotting the number of reviews per business
plt.figure(figsize=(10, 6))
sns.histplot(business_review_counts, bins=30, kde=False)
plt.title('Number of Reviews per Business')
plt.xlabel('Number of Reviews')
plt.ylabel('Count of Businesses')
plt.show()

# Basic statistics for the number of reviews per business
business_review_stats = business_review_counts.describe()
business_review_stats
```

```python
# Calculating the number of reviews written by each user
user_review_counts = df['user_id'].value_counts()

# Plotting the number of reviews per user
plt.figure(figsize=(10, 6))
sns.histplot(user_review_counts, bins=30, kde=False)
plt.title('Number of Reviews per User')
plt.xlabel('Number of Reviews')
plt.ylabel('Count of Users')
plt.show()

# Calculating the number of images posted by each user
# Counting the number of images in the 'pics' column for each user
df['num_pics'] = df['pic_count']
user_image_counts = df.groupby('user_id')['num_pics'].sum()

# Plotting the number of images posted per user
plt.figure(figsize=(10, 6))
sns.histplot(user_image_counts, bins=30, kde=False)
plt.title('Number of Images Posted per User')
plt.xlabel('Number of Images')
plt.ylabel('Count of Users')
plt.show()

# Basic statistics for reviews and images per user
user_review_stats = user_review_counts.describe()
user_image_stats = user_image_counts.describe()

user_review_stats, user_image_stats
```
